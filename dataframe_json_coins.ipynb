```python
import pandas as pd
import json
with open('quotes.json', 'r+', encoding='utf8') as f:
    data = json.load(f)
    for element in data:
        jsonfile = element["teste"]
with open('newquotes.json', 'w') as f:
    json.dump(jsonfile, f)
with open('newquotes.json', 'r+', encoding='utf8') as f:
    dataf = json.load(f)
df=pd.read_json(dataf)
```
